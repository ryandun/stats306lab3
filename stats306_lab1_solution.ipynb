## Control Flow

For people who have programmed in other languages before, here are some examples of R syntax. If you aren't familiar with these, don't worry about it. We will go through some of these later in the course, and much of the course will be learning ways to avoid using this type of code in favor of the high-level data manipulation operations available in `tidyverse`.

```r
# For loops and if-else statements
for(i in 1:10) {
    if (i %% 2 == 0 & i %% 3 == 0) {
        print('FizzBuzz')
    } else if (i %% 2 == 0) {
        print('Fizz')
    } else {
        print(i)
    }
}
```

```r
# Function definitions
exponent = function(x, a=2) {
    x^a  # R automatically returns the last statement
    # return(x^a)  # Explicit returns can also be used
}

exponent(3)
exponent(3, a=3)
```
